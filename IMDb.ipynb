# Classify Movies based on Ratings(Excellent, Good and Average)

```python
def rating(rating):
    if rating>=7.0:
        return "Excellent"
    elif rating>=6.0:
        return "Good"
    else:
        return "Average"
```

```python
df["rating_cat"]=df["Rating"].apply(rating)
```

# Find Unique values from Genre

```python
df['Genre']
```

```python
list1=[]
for i in df["Genre"]:
    list1.append(i.split(","))
list1
```

```python
one_d=[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
one_d
```

```python
uni_list=[]
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
uni_list
```
